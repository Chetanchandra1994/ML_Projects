```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# Read the dataset into python
df = pd.read_csv('YourCabs.csv')
df.head()
```

# Drop id(unique) and user_id(PII)

```python
df = df.drop(['id','user_id'],axis=1)
df.head()
```

```python
df.info()
```

```python
# creating a col to have information for if the VMID was 12 or not
df['is_VMID_12'] = pd.get_dummies(df['vehicle_model_id'].astype('object'))[[12]].astype('int')
df = df.drop(['vehicle_model_id'],axis=1)
df.head()
```

# Dividing the df into 3 parts based on the travel_type_id

```python
df.travel_type_id.unique()
```

```python
df_ttid_1 = df.loc[df.travel_type_id == 1]
df_ttid_1.head()
```

```python
df_ttid_2 = df.loc[df.travel_type_id == 2]
df_ttid_2.head()
```

```python
df_ttid_3 = df.loc[df.travel_type_id == 3]
df_ttid_3.head()
```

# df - travel type = 1

```python
df_ttid_1.head()
```

```python
df_ttid_1.shape
```

```python
df_ttid_1.isnull().sum()
```

```python
df_ttid_1 = df_ttid_1.drop(['travel_type_id','package_id','to_area_id','from_lat','from_long','to_lat','to_long'],axis=1)
df_ttid_1.head()
```

```python
df_ttid_1.isnull().sum()
```

```python
df_ttid_1.info()
```

```python
df_ttid_1 = df_ttid_1.astype({'from_area_id':'object','from_city_id':'object','to_city_id':'object'})
df_ttid_1.info()
```

```python
df_ttid_1.describe(include=['object'])
```

```python
df_ttid_1.from_area_id = df_ttid_1.from_area_id.fillna(393)
df_ttid_1.from_city_id = df_ttid_1.from_city_id.fillna(15)
df_ttid_1.to_city_id = df_ttid_1.to_city_id.fillna(32)
df_ttid_1.isnull().sum()
```

# Create cols with H,M,L,Z cancelations

```python
from_area_id_cancelation_perc = df_ttid_1[['from_area_id','Car_Cancellation']].groupby('from_area_id').mean()['Car_Cancellation']
from_area_id_cancelation_perc.head(20)
```

```python
df_ttid_1['from_area_id_cancelation_perc'] = df_ttid_1['from_area_id'].map(from_area_id_cancelation_perc)
df_ttid_1.head()
```

```python
df_ttid_1.loc[df_ttid_1.from_area_id == 571]
```

```python
df_ttid_1['city_route'] = df_ttid_1['from_city_id'].astype('int').astype('str') + ' -> ' + df_ttid_1['to_city_id'].astype('int').astype('str')
df_ttid_1.head()
```

```python
city_route_cancelation_perc = df_ttid_1[['city_route','Car_Cancellation']].groupby('city_route').mean()['Car_Cancellation']
df_ttid_1['city_route_cancelation_perc'] = df_ttid_1['city_route'].map(city_route_cancelation_perc)
df_ttid_1.head()
```

```python
df_ttid_1['from_area_id_cancelation'] = np.where(df_ttid_1['from_area_id_cancelation_perc'] >= 0.67, 'High',
                                        np.where(df_ttid_1['from_area_id_cancelation_perc'] < 0.33,'Low',
                                         'Medium'))
df_ttid_1['from_area_id_cancelation'].loc[df_ttid_1['from_area_id_cancelation_perc'] == 0.0] = 'Zero'
df_ttid_1.head()      

# np.where((df_ttid_1['from_area_id_cancellation_perc']<0.67) & (df_ttid_1['from_area_id_cancellation_perc'])>=0.33, 'Medium', 'Low'
```

```python
df_ttid_1['city_route_cancelation'] = np.where(df_ttid_1['city_route_cancelation_perc'] >= 0.67, 'High',
                                        np.where(df_ttid_1['city_route_cancelation_perc'] < 0.33,'Low',
                                         'Medium'))
df_ttid_1['city_route_cancelation'].loc[df_ttid_1['city_route_cancelation_perc'] == 0.0] = 'Zero'
df_ttid_1.head()
```

```python
df_ttid_1[['from_area_id','from_area_id_cancelation_perc','from_area_id_cancelation','from_city_id','to_city_id','city_route','city_route_cancelation_perc','city_route_cancelation']].head()
```

```python
df_ttid_1 = df_ttid_1.drop(['from_area_id','from_area_id_cancelation_perc','from_city_id','to_city_id','city_route','city_route_cancelation_perc'],axis=1)
df_ttid_1.head()
```

```python
df_ttid_1.from_date = pd.to_datetime(df_ttid_1.from_date,format='mixed')
df_ttid_1.booking_created = pd.to_datetime(df_ttid_1.booking_created,format='mixed')
df_ttid_1.head()
```

```python
df_ttid_1.info()
```

```python
df_ttid_1['month'] = df_ttid_1.from_date.dt.month
df_ttid_1.head()
```

```python
df_ttid_1['day_of_week'] = df_ttid_1.from_date.dt.dayofweek  # Monday is 0, Sunday is 6
df_ttid_1.head()
```

```python
df_ttid_1['is_weekend'] = (df_ttid_1.from_date.dt.weekday >= 5).astype('int')
df_ttid_1.head()
```

```python
df_ttid_1['hour_of_day'] = df_ttid_1.from_date.dt.hour
df_ttid_1.head()
```

```python
df_ttid_1['time_of_day'] = pd.cut(df_ttid_1['hour_of_day'],bins=[0,6,12,18,24],labels=['Night','Morning','Afternoon','Evening'])
df_ttid_1.head()
```

```python
df_ttid_1['time_diff'] = (df_ttid_1.from_date - df_ttid_1.booking_created).dt.total_seconds()/3600
df_ttid_1.head()
```

```python
df_ttid_1 = df_ttid_1.drop(['hour_of_day','from_date','booking_created'],axis=1)
df_ttid_1.head()
```

```python
# High 1, Medium 3, Low 2, Zero 4  - labelencoder()
# High 1, Medium 2, Low 3, Zero 4  - map()

df_ttid_1['from_area_id_cancelation'] = df_ttid_1['from_area_id_cancelation'].map({'High':1,'Medium':2,'Low':3,"Zero":4})
df_ttid_1['city_route_cancelation'] = df_ttid_1['city_route_cancelation'].map({'High':1,'Medium':2,'Low':3,"Zero":4})
df_ttid_1['time_of_day'] = df_ttid_1['time_of_day'].map({'Morning':1,'Afternoon':2,'Evening':3,"Night":4})
df_ttid_1.head()
```

```python
df_ttid_1.describe()
```

```python
df_ttid_1.boxplot('time_diff')
```

```python
Q1 = df_ttid_1['time_diff'].quantile(0.25)
Q3 = df_ttid_1['time_diff'].quantile(0.75)

IQR = Q3 - Q1

LL = Q1 - 1.5 * IQR
UL = Q3 + 1.5 * IQR

print(f"Q1: {Q1} | Q3: {Q3} | IQR: {IQR} | LL: {LL} | UL: {UL}")
```

```python
outlier_count = df_ttid_1[df_ttid_1['time_diff'] > 97.8].shape[0]

outlier_perc = (outlier_count * 100)/df_ttid_1.shape[0]

print(f"# of Outliers: {outlier_count} | % of Outliers: {outlier_perc}")
```

```python
df_ttid_1['booking_nature'] = pd.cut(df_ttid_1['time_diff'],bins=[-8,9,17,45,1500],labels=['Urgent','SameDay','Regular','Advanced'])
df_ttid_1.head()
```

```python
df_ttid_1 = df_ttid_1.drop('time_diff',axis=1)
df_ttid_1['booking_nature'] = df_ttid_1['booking_nature'].map({'Urgent':1,'SameDay':2,'Regular':3,"Advanced":4})
df_ttid_1.head()
```

```python
df_ttid_1.info()
```

```python
df_ttid_1['time_of_day'] = pd.to_numeric(df_ttid_1['time_of_day'],errors='coerce')
```

```python
df_ttid_1['time_of_day'] = df_ttid_1['time_of_day'].fillna(df_ttid_1['time_of_day'].median())
```

```python
df_ttid_1.info()
```

```python
df_ttid_1['time_of_day'] = df_ttid_1['time_of_day'].astype('int')
df_ttid_1['booking_nature'] = df_ttid_1['booking_nature'].astype('int')
df_ttid_1.info()
```
